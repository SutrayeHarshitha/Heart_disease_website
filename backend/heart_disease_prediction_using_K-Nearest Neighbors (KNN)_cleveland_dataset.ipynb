```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python

# Loading the csv data to a Pandas Dataframe 
heart_data = pd.read_csv('heart_cleveland.csv')
```

```python

# print first 5 rows of the dataset
heart_data.head()
```

```python

# print last 5 rows of the dataset
heart_data.tail()
```

```python

# The number of rows and columns in the dataset 
heart_data.shape
```

```python

# Getting some info about the data 
heart_data.info()
```

```python

# Checking for missing values
heart_data.isnull().sum()
```

```python

# statistical measures about the data 
heart_data.describe()
```

```python

# Checking the distribution of target variable (target values)
# (1) ---------> Defective heart 
# (0) ---------> Healthy heart
heart_data['target'].value_counts()
```

```python

# Splitting the features and target values 
x = heart_data.drop(columns='target',axis=1)
y = heart_data['target']
print(x)
```

```python

print(y)
```

```python

# splitting the data into training data and testing data 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify = y,random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python

# Model initialization
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python

# Training the model
knn_model.fit(x_train, y_train)
```

```python

# Accuracy on training data
x_train_pred_knn = knn_model.predict(x_train)
train_acc_knn = accuracy_score(y_train, x_train_pred_knn)
print("KNN - Accuracy on Training Data: ", train_acc_knn)
```

```python

# Accuracy on testing data
x_test_pred_knn = knn_model.predict(x_test)
test_acc_knn = accuracy_score(y_test, x_test_pred_knn)
print("KNN - Accuracy on Testing Data: ", test_acc_knn)
```
